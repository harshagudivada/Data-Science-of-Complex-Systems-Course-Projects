# Calculating the degree distributions for the first 3 networks

```python
def plot_degree_distribution_undirected_no_weight(G):
  node_degree = pd.Series(dict(G.degree()), name='degree')
  deg = node_degree.value_counts()
  NormDeg = node_degree.value_counts(normalize=True).sort_index()
  degre = list(NormDeg.index)
  prob = list(NormDeg)
  fig = plt.figure(figsize = (10, 5))
  #breakpoint()
 
  # creating the bar plot
  plt.bar(degre, prob, color ='maroon',width = 0.2)

  plt.xlabel("degree")
  plt.ylabel("probability")
  plt.title("Degree Distribution")
  plt.show()
```

```python
plot_degree_distribution_undirected_no_weight(network_4_erdos_renyi)
```

```python
plot_degree_distribution_undirected_no_weight(network_5_erdos_renyi)
```

```python
plot_degree_distribution_undirected_no_weight(network_6_erdos_renyi)
```

```python
def plot_degree_distribution_directed_no_weight(G):
    degrees = [G.degree(n) for n in G.nodes()]
    plt.hist(degrees)
    plt.show()
```

```python
plot_degree_distribution_directed_no_weight(network_1_political)
```
